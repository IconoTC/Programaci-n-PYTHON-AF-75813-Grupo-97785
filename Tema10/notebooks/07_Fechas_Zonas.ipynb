Uso de `datetime`, `zoneinfo` y `calendar` para generar marcas de tiempo y calcular una ventana de mantenimiento.

## 1. Preparar rutas del tema

```python
from pathlib import Path


def localizar_tema10() -> Path:
    """Localiza el directorio Tema10 aunque el notebook se ejecute desde notebooks, Tema10 o Curso_Python."""
    actual = Path.cwd().resolve()
    candidatos = [
        actual,
        actual.parent,
        actual / "Tema10",
        actual.parent / "Tema10",
    ]

    for candidato in candidatos:
        if candidato.name == "Tema10" and candidato.exists():
            return candidato
        if (candidato / "src").exists() and (candidato / "notebooks").exists():
            return candidato

    raise RuntimeError("No se ha podido localizar el directorio Tema10. Abre el notebook desde Tema10/notebooks o desde el proyecto Curso_Python.")


base_dir = localizar_tema10()
src_dir = base_dir / "src"
data_dir = base_dir / "data"
output_dir = base_dir / "output"
backup_dir = base_dir / "backup"
log_dir = base_dir / "logs"

for directorio in (data_dir, output_dir, backup_dir, log_dir):
    directorio.mkdir(parents=True, exist_ok=True)

print("base_dir:", base_dir)
print("data_dir:", data_dir)
```

## 2. Importar módulos de fecha y zona horaria

```python
from datetime import datetime, timezone
from zoneinfo import ZoneInfo
import calendar
```

## 3. Fecha y hora actual en UTC

```python
ahora_utc = datetime.now(timezone.utc)
print("UTC:", ahora_utc.isoformat(timespec="seconds"))
```

## 4. Calcular último día del mes

```python
anio = ahora_utc.year
mes = ahora_utc.month
_, ultimo_dia = calendar.monthrange(anio, mes)
print("Año:", anio)
print("Mes:", mes)
print("Último día:", ultimo_dia)
```

## 5. Ventana de mantenimiento en zona local

```python
zona_local = ZoneInfo("Europe/Madrid")
ejecucion_local = datetime(anio, mes, ultimo_dia, 22, 0, tzinfo=zona_local)
print("Planificación local:", ejecucion_local.isoformat(timespec="seconds"))
```
