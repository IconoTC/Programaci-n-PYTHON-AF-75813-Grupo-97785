# Laboratorio 2. Rutas y ficheros

Uso de `pathlib`, `shutil` y `glob` para preparar directorios, copiar ficheros y buscar por patrón.

## 1. Preparar rutas del tema

```python
from pathlib import Path


def localizar_tema10() -> Path:
    """Localiza el directorio Tema10 aunque el notebook se ejecute desde notebooks, Tema10 o Curso_Python."""
    actual = Path.cwd().resolve()
    candidatos = [
        actual,
        actual.parent,
        actual / "Tema10",
        actual.parent / "Tema10",
    ]

    for candidato in candidatos:
        if candidato.name == "Tema10" and candidato.exists():
            return candidato
        if (candidato / "src").exists() and (candidato / "notebooks").exists():
            return candidato

    raise RuntimeError("No se ha podido localizar el directorio Tema10. Abre el notebook desde Tema10/notebooks o desde el proyecto Curso_Python.")


base_dir = localizar_tema10()
src_dir = base_dir / "src"
data_dir = base_dir / "data"
output_dir = base_dir / "output"
backup_dir = base_dir / "backup"
log_dir = base_dir / "logs"

for directorio in (data_dir, output_dir, backup_dir, log_dir):
    directorio.mkdir(parents=True, exist_ok=True)

print("base_dir:", base_dir)
print("data_dir:", data_dir)
```

## 2. Importar módulos

`pathlib` construye rutas portables, `shutil` copia directorios/ficheros y `glob` localiza ficheros por patrón.

```python
from pathlib import Path
import shutil
import glob
```

## 3. Preparar directorios de datos y backup

```python
data_dir.mkdir(parents=True, exist_ok=True)
backup_dir.mkdir(parents=True, exist_ok=True)
print("Data  :", data_dir)
print("Backup:", backup_dir)
```

## 4. Crear ficheros de ejemplo

```python
(data_dir / "config1.json").write_text('{"servicio": "ssh", "estado": "activo"}', encoding="utf-8")
(data_dir / "config2.json").write_text('{"servicio": "nginx", "estado": "activo"}', encoding="utf-8")
(data_dir / "notas.txt").write_text("Fichero auxiliar", encoding="utf-8")
print("Ficheros creados en:", data_dir)
```

## 5. Copiar el directorio `data` a `backup/data_copia`

```python
destino_copia = backup_dir / "data_copia"
shutil.copytree(data_dir, destino_copia, dirs_exist_ok=True)
print("Copia creada en:", destino_copia)
```

## 6. Buscar ficheros JSON con `glob`

```python
patron = str(destino_copia / "*.json")
archivos = glob.glob(patron)
for archivo in archivos:
    print("JSON respaldado:", archivo)
```
