## 1. Devolver varios valores

Una función puede devolver varios valores separados por comas. Python los agrupa internamente como una tupla.

```python
def resumen_tiempos(valores):
    """
    Calcula el mínimo, máximo y promedio de una lista de tiempos.

    Args:
        valores: Secuencia de tiempos numéricos.

    Returns:
        Una tupla con mínimo, máximo y media.
    """
    minimo = min(valores)
    maximo = max(valores)
    media = sum(valores) / len(valores)
    return minimo, maximo, media
```

## 2. Recuperar varios valores en varias variables

La devolución múltiple permite asignar cada resultado a una variable distinta.

```python
tiempos_respuesta = [12, 9, 15, 11]
mn, mx, avg = resumen_tiempos(tiempos_respuesta)

print("Tiempo mínimo:", mn)
print("Tiempo máximo:", mx)
print("Tiempo medio:", round(avg, 2))
```

## 3. Guardar la devolución completa

También podemos guardar todos los valores devueltos en una sola variable. En ese caso se obtiene una tupla.

```python
resumen = resumen_tiempos([20, 18, 25, 19])
print("Resumen como tupla:", resumen)
print("Tipo:", type(resumen))
```

## 4. Alcance local de variables

La variable `ruta` se crea dentro de la función. Fuera de ella no existe. El dato que debe salir de la función se devuelve con `return`.

```python
def construir_ruta(base, fichero):
    """
    Construye una ruta uniendo directorio base y nombre de fichero.

    Args:
        base: Ruta base.
        fichero: Nombre del fichero.

    Returns:
        Ruta completa.
    """
    ruta = f"{base}/{fichero}"
    return ruta


ruta_log = construir_ruta("/var/log", "app.log")
print("Ruta construida:", ruta_log)
```
