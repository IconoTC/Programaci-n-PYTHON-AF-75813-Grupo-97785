## 1. Función sin parámetros

Esta primera función no recibe datos. Contiene una instrucción interna y muestra un mensaje por pantalla con `print()`.

```python
def saludar_operador():
    """Muestra un mensaje básico de inicio de sesión de soporte."""
    mensaje = "Sesión de soporte iniciada"
    print(mensaje)
```

## 2. Llamada a la función

La función anterior no se ha ejecutado al definirla. Se ejecuta ahora al llamarla.

```python
saludar_operador()
```

## 3. Función con un parámetro

El parámetro `servicio` permite reutilizar la misma función con diferentes valores.

```python
def mostrar_estado(servicio):
    """Muestra por pantalla el nombre de un servicio que se está comprobando."""
    print(f"Comprobando servicio: {servicio}")
```

## 4. Varias llamadas independientes

Cada llamada entra en la función, ejecuta su bloque y vuelve al punto desde el que se llamó.

```python
mostrar_estado("ssh")
mostrar_estado("httpd")
mostrar_estado("backup")
```

## 5. Función que devuelve un valor

`estado_cpu()` no imprime directamente el resultado. Lo calcula y lo devuelve con `return` para que el programa pueda reutilizarlo.

```python
def estado_cpu(uso):
    """
    Clasifica el uso de CPU en un estado operativo.

    Args:
        uso: Porcentaje de uso de CPU.

    Returns:
        Una cadena con el estado NORMAL, ALTO o CRÍTICO.
    """
    if uso >= 90:
        return "CRÍTICO"
    if uso >= 70:
        return "ALTO"
    return "NORMAL"
```

## 6. Prueba con diferentes valores

Modifica `uso_actual` y vuelve a ejecutar la celda para comprobar cómo cambia el resultado.

```python
uso_actual = 82
print(f"Uso de CPU {uso_actual}% -> {estado_cpu(uso_actual)}")

for uso in [35, 72, 94]:
    print(f"Uso de CPU {uso}% -> {estado_cpu(uso)}")
```
