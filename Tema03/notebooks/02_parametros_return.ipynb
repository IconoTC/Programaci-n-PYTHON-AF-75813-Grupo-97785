## 1. Parámetros y argumentos

Los parámetros son los nombres definidos en la cabecera de la función. Los argumentos son los valores concretos enviados cuando la función se llama.

```python
def calcular_iva(base, porcentaje):
    """
    Calcula el importe del IVA aplicado sobre una base.

    Args:
        base: Importe base.
        porcentaje: Porcentaje de IVA.

    Returns:
        Importe del impuesto calculado.
    """
    impuesto = base * porcentaje / 100
    return impuesto
```

## 2. Llamadas a la función

En estas llamadas, `100` y `21` son argumentos. Python los asigna a `base` y `porcentaje` por posición.

```python
iva = calcular_iva(100, 21)
print(f"IVA calculado: {iva:.2f} €")

print(f"IVA reducido: {calcular_iva(250, 10):.2f} €")
```

## 3. Separar cálculo y presentación

La función `calcular_descuento()` devuelve el precio calculado. El `print()` se hace fuera de la función.

```python
def calcular_descuento(precio, descuento):
    """
    Calcula el precio final aplicando un porcentaje de descuento.

    Args:
        precio: Precio inicial.
        descuento: Porcentaje de descuento.

    Returns:
        Precio después de aplicar el descuento.
    """
    return precio - (precio * descuento / 100)


precio_final = calcular_descuento(80, 15)
print(f"Precio final: {precio_final:.2f} €")
```

## 4. Función sin `return` explícito

Si una función no ejecuta `return`, Python devuelve automáticamente `None`. Esto es importante cuando esperamos reutilizar un resultado.

```python
def registrar_evento(texto):
    """
    Muestra un evento por pantalla.

    Esta función no devuelve ningún resultado explícito.
    Python devolverá None automáticamente.
    """
    print(f"EVENTO: {texto}")


resultado = registrar_evento("reinicio programado")
print("Valor devuelto por registrar_evento:", resultado)
```

## 5. Prueba rápida

Modifica los importes y descuentos para comprobar que las funciones siguen funcionando sin tocar su código interno.

```python
precio = 150
descuento = 12

resultado = calcular_descuento(precio, descuento)
print(f"Con descuento del {descuento}%: {resultado:.2f} €")
```
