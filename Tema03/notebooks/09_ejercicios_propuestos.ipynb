## 1. Área de rectángulo

Crear una función `area_rectangulo(base, altura)` que devuelva el área y otra función `mostrar_resultado()` que imprima el mensaje final.

```python
def area_rectangulo(base, altura):
    """Devuelve el área de un rectángulo."""
    # Completar
    return base * altura


def mostrar_resultado(base, altura):
    """Muestra el resultado del cálculo."""
    area = area_rectangulo(base, altura)
    print(f"Área del rectángulo: {area}")


mostrar_resultado(5, 3)
```

## 2. Estado de servicio

Crear una función `estado_servicio(nombre, activo=True)` que devuelva un texto con el nombre del servicio y su estado.

```python
def estado_servicio(nombre, activo=True):
    """Devuelve un texto con el estado del servicio."""
    estado = "activo" if activo else "inactivo"
    return f"{nombre}: {estado}"


print(estado_servicio("ssh"))
print(estado_servicio("backup", activo=False))
```

## 3. Promedio de métricas

Crear una función `promedio(*valores)` que devuelva `None` si no recibe valores y la media si recibe uno o más números.

```python
def promedio(*valores):
    """Devuelve la media de varios valores o None si no recibe datos."""
    if not valores:
        return None
    return sum(valores) / len(valores)


print(promedio(10, 8, 9))
print(promedio())
```

## 4. Normalizar usuario

Crear una función `normalizar_usuario(usuario: str) -> str` que elimine espacios, convierta a minúsculas y sustituya espacios internos por guiones bajos.

```python
def normalizar_usuario(usuario: str) -> str:
    """Normaliza un nombre de usuario."""
    usuario = usuario.strip().lower()
    return usuario.replace(" ", "_")


print(normalizar_usuario("  Ana Sistemas  "))
```

## 5. Resumen de backup

Crear una función `resumen_backup(nombre, *rutas, comprimido=False)` que devuelva un texto indicando el nombre del backup, las rutas incluidas y si se comprimirá.

```python
def resumen_backup(nombre, *rutas, comprimido=False):
    """Devuelve un resumen de backup."""
    if rutas:
        rutas_txt = ", ".join(rutas)
    else:
        rutas_txt = "sin rutas"

    modo = "comprimido" if comprimido else "sin compresión"
    return f"{nombre}: {rutas_txt} -> {modo}"


print(resumen_backup("backup-diario", "/etc", "/var/log", comprimido=True))
print(resumen_backup("backup-vacio"))
```
