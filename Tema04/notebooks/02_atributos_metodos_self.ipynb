## 1. Clase con atributos inicializados desde `__init__()`

`__init__()` se ejecuta al crear la instancia. `self` representa el objeto concreto que se está inicializando.

```python
class Servicio:
    def __init__(self, nombre, activo=False):
        self.nombre = nombre
        self.activo = activo

    def esta_activo(self):
        return self.activo


ssh = Servicio("ssh")
httpd = Servicio("httpd", True)

print(ssh.nombre, ssh.esta_activo())
print(httpd.nombre, httpd.esta_activo())
```

## 2. Añadir métodos que modifican el estado

En el script `.py` estos métodos aparecen comentados para descomentarlos en un segundo paso. En el notebook se muestran en una celda nueva para ejecutarlos de forma limpia.

```python
class Servicio:
    def __init__(self, nombre, activo=False):
        self.nombre = nombre
        self.activo = activo

    def esta_activo(self):
        return self.activo

    def iniciar(self):
        self.activo = True

    def detener(self):
        self.activo = False


ssh = Servicio("ssh")
httpd = Servicio("httpd", True)

ssh.iniciar()
print(ssh.nombre, ssh.esta_activo())
print(httpd.nombre, httpd.esta_activo())

ssh.detener()
print(ssh.nombre, ssh.esta_activo())
```

## 3. Resultado esperado

Solo cambia el objeto sobre el que se invoca el método. `self` permite modificar el estado de esa instancia concreta.
