## 1. Clase mínima con `pass`

`pass` no ejecuta ninguna acción. Sirve para mantener un bloque válido cuando todavía no queremos escribir implementación.

```python
class Servicio:
    pass
```

## 2. Crear instancias

Crear una instancia consiste en llamar a la clase como si fuera una función. Cada llamada crea un objeto independiente.

```python
ssh = Servicio()
web = Servicio()

print(type(ssh))
print(ssh)
print(web)
print(ssh is web)
```

## 3. Comprobación

`ssh is web` devuelve `False` porque son dos objetos distintos, aunque procedan de la misma clase.
