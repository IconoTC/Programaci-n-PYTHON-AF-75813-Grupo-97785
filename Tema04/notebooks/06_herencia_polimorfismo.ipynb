## 1. Clase base y clase hija

La herencia expresa una relación “es un tipo de”. `ServicioWeb` es un tipo especializado de `Servicio`.

```python
class Servicio:
    def __init__(self, nombre, activo=False):
        self.nombre = nombre
        self.activo = activo

    def descripcion(self):
        return f"{self.nombre}: servicio genérico"


class ServicioWeb(Servicio):
    def __init__(self, nombre, puerto, activo=False):
        super().__init__(nombre, activo)
        self.puerto = puerto

    def descripcion(self):
        estado = "activo" if self.activo else "detenido"
        return f"{self.nombre}:{self.puerto} -> {estado}"


web1 = ServicioWeb("nginx", 80, True)
web2 = ServicioWeb("httpd", 8080)

print(web1.descripcion())
print(web2.descripcion())
```

## 2. Polimorfismo básico

Añadimos otra clase hija con el mismo método `descripcion()`. El bucle puede tratar objetos diferentes de forma común.

```python
class ServicioBD(Servicio):
    def __init__(self, nombre, motor, activo=False):
        super().__init__(nombre, activo)
        self.motor = motor

    def descripcion(self):
        return f"{self.nombre} -> motor {self.motor}"


servicios = [
    ServicioWeb("https", 443, True),
    ServicioBD("base-datos", "PostgreSQL", True),
]

for servicio in servicios:
    print(servicio.descripcion())
```

## 3. Herencia múltiple

Python permite herencia múltiple. Aquí se muestra solo para reconocerla; no debe ser la primera opción de diseño.

```python
class Monitorizable:
    def comprobar(self):
        return "comprobado"


class Reiniciable:
    def reiniciar(self):
        return "reiniciado"


class ServicioSistema(Monitorizable, Reiniciable):
    pass


svc = ServicioSistema()
print(svc.comprobar())
print(svc.reiniciar())
```
