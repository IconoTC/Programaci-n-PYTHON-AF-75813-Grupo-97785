## 1. Imprimir una instancia sin `__str__()`

Sin `__str__()`, Python muestra una representación técnica del objeto.

```python
class Servicio:
    def __init__(self, nombre, puerto):
        self.nombre = nombre
        self.puerto = puerto


MiServicio1 = Servicio("web", 80)
print(MiServicio1)
print(f"Servicio registrado -> {MiServicio1.nombre} -> {MiServicio1.puerto}")
```

## 2. Añadir `__str__()`

`__str__()` debe devolver texto con `return`, no imprimir directamente. Así el objeto se puede usar en `print()`, f-strings o logs.

```python
class Servicio:
    def __init__(self, nombre, puerto):
        self.nombre = nombre
        self.puerto = puerto

    def __str__(self):
        return f"{self.nombre}:{self.puerto}"


MiServicio1 = Servicio("web", 80)
print(MiServicio1)
print(f"Servicio registrado -> {MiServicio1.nombre} -> {MiServicio1.puerto}")
```

## 3. Atributo de clase y atributo de instancia

`categoria` es un atributo de clase. Se comparte entre instancias salvo que se sobrescriba en una instancia concreta.

```python
class Servicio:
    categoria = "Sistema"

    def __init__(self, nombre, puerto):
        self.nombre = nombre
        self.puerto = puerto

    def __str__(self):
        return f"{self.nombre}:{self.puerto}"


MiServicio1 = Servicio("web", 80)
MiServicio2 = Servicio("ssh", 22)

print(MiServicio1)
print(MiServicio2)
print(MiServicio1.categoria, MiServicio2.categoria)

print("Cambiamos categoria en MiServicio1")
MiServicio1.categoria = "aplicacion"
print(MiServicio1.categoria, MiServicio2.categoria)
```

## 4. Observación

Al asignar `MiServicio1.categoria`, se crea o sobrescribe ese atributo en la instancia. `MiServicio2` sigue leyendo el atributo de clase.
