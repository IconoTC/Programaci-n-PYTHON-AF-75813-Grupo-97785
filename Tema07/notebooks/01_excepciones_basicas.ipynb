## 2. `try/except` básico

Se captura únicamente el error esperado: `ValueError`.

```python
valor = "ssh"

try:
    puerto = int(valor)
    print("Puerto convertido:", puerto)
except ValueError:
    print(f"No se puede convertir {valor!r} a número entero.")
```

## 3. Capturar el objeto de excepción con `as`

`as error` permite mostrar un mensaje controlado y, si resulta útil, un detalle técnico.

```python
valor = "https"

try:
    puerto = int(valor)
    print("Puerto convertido:", puerto)
except ValueError as error:
    print("Puerto no válido.")
    print("Detalle técnico:", error)
```

## 4. Excepciones frecuentes en estructuras de datos

`KeyError` aparece al solicitar una clave inexistente. `IndexError` aparece al solicitar una posición fuera de rango.

```python
datos = {"ssh": 22}
servicios = ["ssh", "nginx"]

try:
    print("Puerto de api:", datos["api"])
except KeyError as error:
    print("Clave inexistente en diccionario:", error)

try:
    print("Servicio en posición 5:", servicios[5])
except IndexError as error:
    print("Índice fuera de rango:", error)
```

## 5. División entre cero

Se captura `ZeroDivisionError` porque el divisor puede valer cero.

```python
dividendo = 10
divisor = 0

try:
    resultado = dividendo / divisor
    print(resultado)
except ZeroDivisionError:
    print("No se puede dividir por cero.")
```

## 6. Capturar varias excepciones con la misma respuesta

Cuando la respuesta del programa es la misma, varias excepciones pueden agruparse en una tupla.

```python
datos = ["22", "abc"]
indice = 2

try:
    puerto = int(datos[indice])
    print("Puerto:", puerto)
except (ValueError, IndexError) as error:
    print("No se pudo obtener el puerto.")
    print("Tipo:", type(error).__name__)
    print("Detalle:", error)
```

## 7. Bloque `try` pequeño y centrado

El `try` debe envolver solo la operación que puede fallar, no todo el flujo del programa.

```python
valor = "443"
puerto = None

try:
    puerto = int(valor)
except ValueError:
    print("Dato no convertible.")

if puerto is not None:
    print("Puerto convertido correctamente:", puerto)
```
