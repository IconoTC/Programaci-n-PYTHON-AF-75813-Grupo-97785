Diagnóstico básico de importaciones.

## 1. Objetivo

Revisar errores frecuentes al importar módulos y observar que Python genera el directorio `__pycache__` al importar módulos.

## 2. Forzar una importación para generar `__pycache__`

La celda ejecuta código Python completo dentro del heredoc.

```python
%%bash
cd ../src
python - <<'PY'

import validaciones

print("Puerto 443 válido:", validaciones.puerto_valido(443))
PY
```

## 3. Comprobar el directorio `__pycache__`

Esta comprobación usa comandos de terminal para listar el directorio creado automáticamente.

```python
%%bash
cd ../src
ls -la
if [ -d "__pycache__" ]; then
    echo ""
    echo "Contenido de __pycache__:"
    ls -la __pycache__
else
    echo "No existe __pycache__"
fi
```

## 4. Ejecutar el código completo de errores de importación

La celda contiene el código de `06_errores_importacion.py` dentro del heredoc.

```python
%%bash
cd ../src
python - <<'PY'
"""
Tema 9 - Módulos
Laboratorio 6: errores frecuentes al importar.

Objetivo:
    Identificar errores comunes de importación sin romper el laboratorio.
"""

print("=== Error 1: recurso inexistente ===")

try:
    from validaciones import validar_ip
except ImportError as error:
    print("ImportError controlado:")
    print(error)

print("\n=== Error 2: módulo inexistente ===")

try:
    import modulo_que_no_existe
except ModuleNotFoundError as error:
    print("ModuleNotFoundError controlado:")
    print(error)

print("\n=== Diagnóstico recomendado ===")
print("- Comprobar el nombre del fichero .py")
print("- Comprobar que estamos ejecutando desde el directorio correcto")
print("- Revisar si el recurso existe dentro del módulo")
print("- Evitar nombres de fichero como random.py, json.py o sys.py")
print("- Revisar si dos módulos se importan entre sí")
PY
```

## 5. Interpretación

`ImportError` aparece cuando el módulo existe, pero el recurso solicitado no. `ModuleNotFoundError` aparece cuando Python no encuentra el módulo solicitado.
