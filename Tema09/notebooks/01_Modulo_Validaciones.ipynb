## 1. Objetivo

Crear y probar el módulo `validaciones.py`. Este fichero no lleva numeración porque debe poder importarse con `import validaciones`.

## 2. Probar el módulo desde una ejecución externa

La celda ejecuta Python desde el directorio `src` y coloca **todo el código de prueba dentro del heredoc**. Así se ve el código completo que se ejecuta y se reproduce el comportamiento de una terminal.

```python
%%bash
cd ../src
python - <<'PY'

import validaciones

print(validaciones.puerto_valido(443))
print(validaciones.normalizar_servicio(" SSH "))
print(validaciones.estado_valido(" ok "))
PY
```

## 3. Punto clave

El fichero se llama `validaciones.py`, por eso puede importarse con `import validaciones`. Si se llamara de otro modo, la importación normal fallaría.
