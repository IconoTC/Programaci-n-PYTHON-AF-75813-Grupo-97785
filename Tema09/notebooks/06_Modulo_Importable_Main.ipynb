# Tema 9 - Laboratorio 6: módulo importable y script ejecutable

Uso práctico de `__name__` y `__main__`.

## 1. Objetivo

Diferenciar un módulo importable de un script ejecutable mediante `if __name__ == "__main__"`.

## 2. Ejecutar directamente el módulo `diagnostico.py`

La celda contiene el código completo de `diagnostico.py`. Al ejecutarse como programa principal, el bloque `if __name__ == "__main__"` sí se ejecuta.

```python
%%bash
cd ../src
python - <<'PY'
"""
Tema 9 - Módulos
Módulo de diagnóstico.

Este ejemplo documenta el uso de if __name__ == "__main__".
"""


def comprobar() -> str:
    """Devuelve un mensaje sencillo de diagnóstico."""
    return "sistema operativo OK"


def obtener_estado_servicios() -> list[dict]:
    """Devuelve una lista simulada de servicios."""
    return [
        {"nombre": "ssh", "estado": "OK"},
        {"nombre": "nginx", "estado": "OK"},
        {"nombre": "api", "estado": "ERROR"},
    ]


def resumen_servicios(servicios: list[dict]) -> str:
    """Genera un resumen simple de servicios."""
    total = len(servicios)
    errores = 0

    for servicio in servicios:
        if servicio["estado"] == "ERROR":
            errores += 1

    return f"Servicios revisados: {total}. Servicios con error: {errores}"


# Este bloque solo se ejecuta si lanzamos:
# python diagnostico.py
#
# No se ejecuta cuando otro script hace import diagnostico.
if __name__ == "__main__":
    print("Iniciando diagnóstico")
    print(comprobar())

    servicios = obtener_estado_servicios()
    print(resumen_servicios(servicios))
PY
```

## 3. Ejecutar el script que importa `diagnostico.py`

La celda contiene el código completo de `05_main_diagnostico.py`. Al importar `diagnostico`, el bloque `__main__` del módulo no se ejecuta automáticamente.

```python
%%bash
cd ../src
python - <<'PY'
"""
Tema 9 - Módulos
Laboratorio 5: importar un módulo con bloque __main__.

Objetivo:
    Comprobar que el bloque if __name__ == "__main__"
    no se ejecuta cuando el fichero se importa como módulo.
"""

import diagnostico

print("Comprobación:", diagnostico.comprobar())

servicios = diagnostico.obtener_estado_servicios()
print("Servicios:", servicios)

resumen = diagnostico.resumen_servicios(servicios)
print("Resumen:", resumen)
PY
```

## 4. Interpretación

El bloque `if __name__ == "__main__"` sirve para incluir pruebas o ejecución local sin provocar efectos laterales cuando el fichero se importa como módulo.
