## 1. Crear y consultar diccionarios

Un diccionario almacena pares clave-valor. Se accede por clave, no por índice.

```python
servidor = {
    "hostname": "srv-web-01",
    "ip": "10.0.0.20",
    "activo": True,
}

print(servidor)
print(type(servidor))
print("Hostname:", servidor["hostname"])
print("IP:", servidor["ip"])
```

## 2. Acceso seguro con `get()`

`get()` evita `KeyError` cuando una clave no existe y permite indicar un valor por defecto.

```python
servidor = {"hostname": "srv01", "ip": "10.0.0.20"}

print(servidor["hostname"])
print("Ubicación:", servidor.get("ubicacion"))
print("Ubicación con defecto:", servidor.get("ubicacion", "sin asignar"))

if servidor.get("ip"):
    print(f"IP detectada: {servidor['ip']}")

# Descomenta para comprobar KeyError:
# print(servidor["ubicacion"])
```

## 3. Añadir, modificar y eliminar datos

Los diccionarios son mutables. Podemos añadir claves, modificar valores y eliminar entradas.

```python
servicio = {"nombre": "nginx", "puerto": 80}
print("Inicial:", servicio)

servicio["activo"] = True
servicio["puerto"] = 443
print("Tras añadir/modificar:", servicio)

servicio.update({"protocolo": "tcp", "seguro": True})
print("Tras update:", servicio)

puerto = servicio.pop("puerto")
print("Tras pop:", servicio)
print("Puerto extraído:", puerto)

del servicio["seguro"]
print("Tras del:", servicio)
```

## 4. Recorrer claves, valores e items

Al recorrer directamente un diccionario se obtienen claves. `items()` es útil cuando necesitamos clave y valor.

```python
servicio = {
    "nombre": "nginx",
    "puerto": 443,
    "activo": True,
}

for clave in servicio:
    print(clave, servicio[clave])

print()

for clave in servicio.keys():
    print("Clave:", clave)

print()

for valor in servicio.values():
    print("Valor:", valor)

print()

for clave, valor in servicio.items():
    print(f"{clave}: {valor}")
```

## 5. Diccionarios anidados

Permiten representar inventarios y configuraciones con más estructura.

```python
inventario = {
    "srv-web-01": {
        "ip": "10.0.0.20",
        "servicios": ["ssh", "nginx"],
        "entorno": "prod",
    },
    "srv-db-01": {
        "ip": "10.0.0.30",
        "servicios": ["ssh", "postgresql"],
        "entorno": "prod",
    },
}

print("IP de srv-web-01:", inventario["srv-web-01"]["ip"])

for host, datos in inventario.items():
    print(f"{host} -> {datos['ip']} -> {datos['servicios']}")
```

## 6. Diccionarios de objetos

El valor de un diccionario puede ser una instancia de una clase.

```python
class Servicio:
    """Representa un servicio con nombre y estado."""

    def __init__(self, nombre, activo):
        self.nombre = nombre
        self.activo = activo

    def comprobar(self):
        """Devuelve el estado del servicio."""
        if self.activo:
            return "OK"
        return "DETENIDO"


servicios = {
    "ssh": Servicio("ssh", True),
    "api": Servicio("api", False),
}

print("ssh:", servicios["ssh"].comprobar())
print("api:", servicios["api"].comprobar())
```

## 7. Comprensión de diccionarios

Crea pares clave-valor a partir de una colección.

```python
servicios = ["ssh", "nginx", "postgresql"]
puertos = [22, 80, 5432]

mapa_puertos = {servicios[i]: puertos[i] for i in range(len(servicios))}
print("Mapa de puertos:", mapa_puertos)

longitudes = {servicio: len(servicio) for servicio in servicios}
print("Longitudes:", longitudes)
```

## 8. Ejemplo adicional: contar estados

Un diccionario también puede usarse como contador simple.

```python
resultados = ["OK", "FALLO", "OK", "REVISAR", "FALLO", "OK"]
conteo = {}

for estado in resultados:
    if estado not in conteo:
        conteo[estado] = 0
    conteo[estado] += 1

print("Conteo de estados:", conteo)
```
