## Ejercicio 1. Normalizar usuario
Partir de un nombre con espacios, limpiarlo, pasarlo a minúsculas y mostrarlo con una f-string.

```python
usuario_original = "   ADMIN.SOPORTE   "
usuario_normalizado = usuario_original.strip().lower()
print(f"Usuario normalizado: {usuario_normalizado}")
```

## Ejercicio 2. Cálculo de inventario
Calcular el total con IVA y mostrar el resultado redondeado a dos decimales.

```python
unidades_texto = "8"
precio_texto = "24.95"
iva = 1.21

unidades = int(unidades_texto)
precio_unitario = float(precio_texto)
total = unidades * precio_unitario * iva

print(f"Total con IVA: {total:.2f} €")
```

## Ejercicio 3. Clasificar ticket
Cambiar `codigo` y observar la rama ejecutada.

```python
codigo = "u".lower()

match codigo:
    case "b":
        print("Prioridad baja")
    case "m":
        print("Prioridad media")
    case "a":
        print("Prioridad alta")
    case "u":
        print("Prioridad urgente")
    case _:
        print("Prioridad no reconocida")
```

## Ejercicio 4. Validar IP simple
Separar una dirección IP por puntos, comprobar que tiene cuatro partes y mostrar cada octeto.

```python
ip = "192.168.1.50"
partes = ip.split(".")

if len(partes) == 4:
    print("Formato básico correcto")
    for indice, octeto in enumerate(partes):
        print(f"Octeto {indice + 1}: {octeto}")
else:
    print("Formato incorrecto: la IP no tiene cuatro partes")
```

## Ejercicio 5. Reintento de clave
Simular tres intentos de acceso usando una lista. Cambiar `intentos` para probar éxito, fallo y bloqueo.

```python
clave_correcta = "admin123"
intentos = ["test", "root", "admin123"]
max_intentos = 3

for numero, clave in enumerate(intentos[:max_intentos], start=1):
    print(f"Intento {numero}: {clave}")
    if clave == clave_correcta:
        print("Acceso concedido")
        break
else:
    print("Acceso bloqueado")
```
