## Cadena base del laboratorio
Esta cadena se usará en los bloques siguientes para practicar métodos de `str`.

```python
mensaje: str = "servidor ACTIVO en la red local"
print(mensaje)
```

## Transformaciones de formato
Estos métodos devuelven una cadena nueva. La cadena original no cambia salvo que reasignemos el resultado.

```python
print(mensaje.capitalize())
print(mensaje.title())
print(mensaje.upper())
print(mensaje.lower())
print(mensaje.swapcase())
```

## Comprobaciones de contenido
Devuelven `True` o `False`. Sirven para prevalidar texto antes de convertirlo o clasificarlo.

```python
print("mensaje", "isalnum:", mensaje.isalnum())
print("Cisco123", "isalnum:", 'Cisco123'.isalnum())
print("Admin", "isalpha:", 'Admin'.isalpha())
print("2026", "isdigit:", '2026'.isdigit())
print("Vlan10", "isdigit:", 'Vlan10'.isdigit())
print("ROOT_01", "isupper:", 'ROOT_01'.isupper())
print("admin 80", "islower:", 'admin 80'.islower())
```

## Funciones generales aplicadas a cadenas
`len()`, `min()` y `max()` son funciones generales, no métodos de `str`.

```python
print("Caracteres totales:", len(mensaje))
print("Valor máximo ASCII:", max(mensaje))
print("Valor mínimo ASCII:", min(mensaje))
```

## Limpieza de extremos
`lstrip()`, `rstrip()` y `strip()` eliminan espacios o caracteres de los extremos de la cadena, no del centro.

```python
log_entrada = "      Error de sistema      "
print(log_entrada, end="[FIN]\n")
print("lstrip:", log_entrada.lstrip(), end="[FIN]\n")
print("rstrip:", log_entrada.rstrip(), end="[FIN]\n")
print("strip:", log_entrada.strip(), end="[FIN]\n")
print("strip de caracter -:", '---LOG-FILE---'.strip('-'))
```

## Búsqueda, sustitución, división y unión
Estas operaciones son habituales en logs, rutas, direcciones IP y cadenas procedentes de sistemas externos.

```python
print("replace e -> 3:", mensaje.replace("e", "3"))
print("replace a -> 4 solo la 1º a:", mensaje.replace("a", "4", 1))
print("find:", mensaje.find('e'))
print("rfind:", mensaje.rfind('e'))
print("find (no existe):", mensaje.find('z'))
print("count 'a':", mensaje.count('a'))

IP = "192.168.1.50"
IP_split = IP.split(".")
print("IP_split:", IP_split)

partes_ip = ["10", "0", "0", "1"]
print("join:", ".".join(partes_ip))
```
