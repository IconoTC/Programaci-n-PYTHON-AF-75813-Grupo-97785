## Enteros y sensibilidad a mayúsculas
Python no exige declarar el tipo. El tipo queda asociado al objeto asignado. Además, diferencia mayúsculas y minúsculas.

```python
stock_portatiles = 15
# Stock_portatiles = 20  # Python es case sensitive; sería otra variable distinta
stock_monitores = 7

total_equipos = stock_portatiles + stock_monitores
print("Tipo de total_equipos:", type(total_equipos))
print(total_equipos)
```

## Concatenación frente a `print()` con comas
La concatenación con `+` exige cadenas. `print()` con comas acepta distintos tipos y los muestra separados por espacios.

```python
print("Total productos: " + str(total_equipos))
print("Total productos:", total_equipos)
```

## Números reales (`float`)
Se usan para representar valores con decimales.

```python
precio_teclado = 25.99
iva_aplicado = 1.21
precio_final = precio_teclado * iva_aplicado

print("Tipo de precio_final:", type(precio_final))
print("Coste total:", precio_final)
```

## Booleanos (`bool`)
Solo tienen dos valores: `True` o `False`.

```python
envio_gratis = False
print("¿Tiene envío gratuito?", envio_gratis)
print("Tipo de envio_gratis:", type(envio_gratis))
```

## Cadenas de texto (`str`)
Pueden escribirse con comillas simples o dobles.

```python
marca = "Logitech"
modelo = 'G-Pro'

print(marca, modelo)
print(marca + " " + modelo)
print("Tipo de marca:", type(marca))
```
