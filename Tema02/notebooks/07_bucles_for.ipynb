## Recorrido de una lista
Un `for` en Python recorre elementos de un iterable.

```python
componentes = ['CPU', 'GPU', 'RAM', 'SSD', 'PSU']
print("Lista de componentes:", componentes)

print("Presentación en una línea")
for item in componentes:
    print(item, end=" ")
print()

print("Presentación con salto de línea")
for item in componentes:
    print(item)
```

## enumerate()
Permite obtener índice y valor a la vez.

```python
for indice, nombre in enumerate(componentes):
    print(f"ID {indice} -> {nombre}")
```

## f-string frente a concatenación
La f-string suele ser más legible que convertir y concatenar manualmente.

```python
print("Listado con formatted string:")
for i in range(len(componentes)):
    print(f"Posición {i}: {componentes[i]}")

print("Listado concatenando cadenas:")
for i in range(len(componentes)):
    print("Posición " + str(i) + ": " + componentes[i])
```

## range() con inicio, fin y paso
Recuerda que el valor final no se incluye.

```python
print("Rango 0 a 9")
for valor in range(10):
    print(valor, end=" ")
print()

print("Rango 5 a 15")
for valor in range(5, 16):
    print(valor, end=" ")
print()

print("Pares de 10 a 20")
for valor in range(10, 21, 2):
    print(valor, end=" ")
print()

print("Cuenta atrás")
for valor in range(5, 0, -1):
    print(valor, end=" ")
print()
```

## Acumulación de valores
Compara una solución con condicional dentro del bucle y otra usando el parámetro `step` de `range()`.

```python
ventas = [120, 450, 300, 210, 580, 600, 150]

total_impares = 0
for i in range(len(ventas)):
    if i % 2 != 0:
        total_impares += ventas[i]
print("Total acumulado (índices impares):", total_impares)

total_impares = 0
for i in range(1, len(ventas), 2):
    total_impares += ventas[i]
print("Total acumulado (optimizado):", total_impares)
```
