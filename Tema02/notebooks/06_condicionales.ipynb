## Adaptación del uso de `input()`
En el script `.py`, la temperatura y la prioridad se leen por teclado. En el notebook se usan variables editables para ejecutar cada bloque sin esperar entrada interactiva. Cambia `grados` y `codigo` para probar otros caminos.

## Temperatura de prueba

```python
grados = 28
print("Temperatura usada en la prueba:", grados)
```

## Condicionales anidados
Funcionan, pero pueden hacerse difíciles de leer si aumentan las ramas.

```python
if grados < 18:
    print("Alerta: Temperatura demasiado baja")
else:
    if grados > 27:
        print("Alerta: Riesgo de sobrecalentamiento")
    else:
        print("Estado del sistema: Óptimo")
```

## Estructura if / elif / else
Suele ser más legible para rangos o alternativas excluyentes.

```python
if grados < 18:
    print("Alerta: Temperatura demasiado baja")
elif grados > 27:
    print("Alerta: Riesgo de sobrecalentamiento")
else:
    print("Estado del sistema: Óptimo")
```

## Forma abreviada
Existe, pero no siempre mejora la lectura. Se incluye para reconocerla, no como estilo preferente.

```python
if grados < 18: print("Alerta: Temperatura demasiado baja")
elif grados > 27: print("Alerta: Riesgo de sobrecalentamiento")
else: print("Estado del sistema: Óptimo")
```

## match/case para clasificar prioridades
Modifica `codigo` con valores como `b`, `m`, `a`, `u`, `1` o cualquier otro texto.

```python
codigo = "a".lower()

match codigo:
    case 'b':
        print("Prioridad: BAJA. Respuesta en 48 horas.")
    case 'm':
        print("Prioridad: MEDIA. Respuesta en 24 horas.")
    case 'a':
        print("Prioridad: ALTA. Respuesta en 4 horas.")
    case 'u':
        print("Prioridad: CRÍTICA. Respuesta inmediata.")
    case '1' | '2' | '3':
        print("Código numérico detectado. Por favor, use letras.")
    case _:
        print("Código de prioridad no reconocido.")
```
