## Adaptación del uso de `input()`
El script `.py` pide una clave con `input()` en cada intento. En notebook se simula una secuencia de intentos con una lista, para que el bucle sea reproducible y no bloquee VS Code. Cambia la lista `intentos_prueba` para probar otros escenarios.

## Validación de acceso con intentos simulados
El bucle conserva la lógica original: máximo 3 intentos y `break` cuando se acierta.

```python
intentos_realizados = 0
max_intentos = 3
intentos_prueba = ["error", "admin123"]

for acceso in intentos_prueba:
    if intentos_realizados >= max_intentos:
        break

    print("Intento recibido:", acceso)

    if acceso == "admin123":
        print("Acceso concedido al sistema.")
        break

    intentos_realizados += 1
    intentos_restantes = max_intentos - intentos_realizados

    if intentos_restantes > 0:
        print(f"Código erróneo. Le quedan {intentos_restantes} intentos.")
    else:
        print("Acceso bloqueado. Agotó todos los intentos.")
else:
    if intentos_realizados < max_intentos:
        print("No se han proporcionado más intentos de prueba.")
```

## Versión con while sin `input()`
Esta versión mantiene explícitamente el `while`, pero toma los valores desde una lista.

```python
intentos_realizados = 0
max_intentos = 3
intentos_prueba = ["111", "222", "admin123"]
indice = 0

while intentos_realizados < max_intentos and indice < len(intentos_prueba):
    acceso = intentos_prueba[indice]
    indice += 1
    print("Intento recibido:", acceso)

    if acceso == "admin123":
        print("Acceso concedido al sistema.")
        break

    intentos_realizados += 1
    intentos_restantes = max_intentos - intentos_realizados

    if intentos_restantes > 0:
        print(f"Código erróneo. Le quedan {intentos_restantes} intentos.")
    else:
        print("Acceso bloqueado. Agotó todos los intentos.")
```

## Uso de continue
`continue` no termina el bucle: salta el resto de la iteración actual y vuelve a evaluar la condición.

```python
contador = 0
while contador < 5:
    contador += 1
    if contador == 3:
        continue
    print(f"Procesando paso: {contador}")
```
