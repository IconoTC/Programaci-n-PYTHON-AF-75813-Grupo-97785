## Conversión de texto a entero
`input()` siempre devuelve `str`. Para operar aritméticamente con el dato leído hay que convertirlo al tipo correcto.

```python
valor_leido = 15 # Usamos un valor asignado en lugar de input para que no bloquee al notebook. Puedes cambiarlo
# valor_leido = input("Introduce la cantidad de productos (número entero): ") # Se comenta esta linea
print("Tipo antes de la conversión:", type(valor_leido))
cantidad = int(valor_leido)
print("Tipo tras conversión:", type(cantidad))
precio_unidad = 20 # Se usa esta linea para no bloquear. Puedes cambiarlo
# precio_unidad = int(input("Introduce el precio (número entero): ")) # Se comenta esta linea para no bloquear
total_compra = cantidad * precio_unidad
print("Coste total entero:", total_compra)
```

## Conversión de texto a decimal
`float()` permite convertir texto numérico con decimales. `round()` permite redondear la salida.

```python
import math
distancia = 4.56 # Se usa este valor fijo y puedes cambiarlo
# distancia = float(input("Introduce la distancia en km: ")) # Se comenta esta linea para no bloquear
resultado_con_math_pow = math.pi * math.pow(distancia, 2)
resultado_con_operador = math.pi * distancia ** 2
print("Resultado con math.pow:", resultado_con_math_pow)
print("Resultado con operador **:", resultado_con_operador)
print("Resultado redondeado:", round(resultado_con_operador, 3))
```

## Conversión a booleanos
En números, `0` se evalúa como `False` y cualquier otro valor como `True`. En cadenas, cualquier texto no vacío se evalúa como `True`, incluso la cadena `"False"`.

```python
activado = 0
print("¿Está el sistema activo?:", bool(activado))
activado = 1
print("¿Está el sistema activo?:", bool(activado))
estado_texto = "False"
print("Resultado booleano de la cadena 'False':", bool(estado_texto))
```

## Conversión a cadenas
`str()` permite convertir otros tipos a texto para concatenar o mostrar información.

```python
puntuacion = 95
print("Tipo antes de conversión:", type(puntuacion))
mensaje_final = "Tu nota es: " + str(puntuacion)
print(mensaje_final)
```
