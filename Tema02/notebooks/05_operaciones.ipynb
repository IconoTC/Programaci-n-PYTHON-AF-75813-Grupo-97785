## Operaciones aritméticas básicas
Permiten calcular valores numéricos. Observa la diferencia entre división decimal `/`, división entera `//` y resto `%`.

```python
unidades = 9
cajas = 4

print("Suma total:", unidades + cajas)
print("Diferencia:", unidades - cajas)
print("Producto:", unidades * cajas)
print("Cociente decimal:", unidades / cajas)
print("Cociente entero:", unidades // cajas)
print("Resto de división entero:", unidades % cajas)
print("Exponencial:", unidades ** cajas)
```

## Operadores de asignación compuesta
Actualizan una variable usando su valor anterior. Son habituales en contadores y acumuladores.

```python
stock = 20
print("Stock:", stock)

stock += 10
print("Stock:", stock)

stock *= 10
print("Stock:", stock)
```

## Operadores relacionales y lógicos
Los relacionales devuelven booleanos. Los lógicos combinan condiciones.

```python
print("¿Es mayor?:", stock > unidades)
print("¿Es mayor o igual?:", stock >= 50)
print("¿Es idéntico valor?:", stock == 30)
print("¿Es valor distinto?:", stock != 10)

print("Uso de and:", stock > 0 and unidades == 9)
print("Uso de or:", stock < 0 or unidades > 5)
print("Uso de not:", not unidades == 5)
```

## Identidad, igualdad y pertenencia
`==` compara contenido. `is` compara si dos referencias apuntan al mismo objeto. `in` comprueba pertenencia.

```python
carrito_a = ['Monitor', 'Teclado']
carrito_b = ['Monitor', 'Teclado']

print("¿Mismo contenido?:", carrito_a == carrito_b)
print("¿Misma ubicación de lista?:", carrito_a is carrito_b)

usuarios_pro = ['Admin', 'Editor']
print("¿Admin en lista?:", 'Admin' in usuarios_pro)
print("¿Socio fuera de lista?:", 'Socio' not in usuarios_pro)

puntuacion = 85
print("¿Está en rango 0-100?:", 0 <= puntuacion <= 100)
```
